# PREPARING OF DATA FOR PREPARING MODEL 

```python
#10. CREATING DATA FRAME

df = df.head(len(df)-1)
df
```

```python
#11. create empty list

x=[]
y=[]
```

```python
#12. prepare the data for training

df_days=df.loc[:493, 'Date']
df_adj_close = df.loc[:493, 'Average Price']
```

```python
df_days
```

```python
#13. Create the independent data set

for day in df_days:
    #day=day.split('-')
    #x.append(int(day[0]))
    x.append( [int(day.split('-')[0])] )

#14. Create the dependent data set

for adj_close_price in df_adj_close:
    y.append( float(adj_close_price) )
    
```

# SELECTING MODEL FOR THE PROJECT

```python
#15. splitting the data into training and testing set

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
y_train
```

```python
y_test
```

```python
#16. USING LINEAR VERSION OF KERNAL OF SVR

from sklearn.svm import SVR

lin_svr=SVR(kernel='linear', C=1000.0)
lin_svr.fit(x_train,y_train)
```

```python
#17. USING POLYNOMIAL VERSION OF KERNAL OF SVR 

poly_svr=SVR(kernel='poly', degree=2 )
poly_svr.fit(x_train,y_train)

```

```python
#18. USING RBF VERSION OF KERNAL OF SVR

rbf_svr=SVR(kernel='rbf', gamma=0.15)
rbf_svr.fit(x_train,y_train)
```

# PLOTTING THE PREDICTION OF THE MODELS

```python
#19. plot the data predicted by the 3 models
#Plot the models on a graph to see which has the best fit

plt.figure(figsize=(12,6))
plt.scatter(x_train, y_train, color = 'black', label='Original Data')
plt.plot(x, rbf_svr.predict(x), color = 'green', label='RBF Model')
plt.plot(x, poly_svr.predict(x), color = 'orange', label='Polynomial Model')
plt.plot(x, lin_svr.predict(x), color = 'purple', label='Linear Model')
plt.xlabel('Days')
plt.ylabel('Average Price')
plt.title('Support Vector Regression')
plt.legend()
plt.show()
```

```python
#20. PREDICTION OF PRICES FOR THE 30TH DAY  OF THE MONTH

day = [[30]]
print('The RBF SVR predicted:', rbf_svr.predict(day))
print('The Linear SVR predicted:', lin_svr.predict(day))
print('The Polynomial SVR predicted:', poly_svr.predict(day))
```

# Evaluating a model with the score method

```python
#21. ACCURACY OF ALL 3 MODELS WE HAVE USED

print("The LINEAR SVR predicted:",lin_svr.score(x_train,y_train))
print("The POLY SVR predicted:",poly_svr.score(x_train,y_train))
print("The RBF SVR predicted:",rbf_svr.score(x_train,y_train))
```

# Save a model and load it

```python
# 22. Save a model and load it
import pickle

pickle.dump(lin_svr, open("LINEAR SVR.pkl", "wb"))
pickle.dump(poly_svr, open("POLY SVR.pkl", "wb"))
pickle.dump(rbf_svr, open("RBF SVR.pkl", "wb"))
```

```python
loaded_model = pickle.load(open("LINEAR SVR.pkl", "rb"))
loaded_model.score(x_test, y_test)
```
